```python
import gdsfactory as gf

class MZIWithBend:
    """
    Class to create a Mach-Zehnder Interferometer (MZI) with a circular bend.
    """
    def __init__(self, radius: float = 10):
        """
        Initializes the MZIWithBend class.

        Args:
            radius: Radius of the circular bend.
        """
        self.radius = radius
        self.component = gf.Component()

    def create_component(self) -> gf.Component:
        """
        Creates the MZI with a circular bend.

        Returns:
            A gdsfactory Component containing the MZI with a circular bend.
        """
        mzi = self.component << gf.components.mzi(delta_length=10)
        bend = self.component << gf.components.bend_circular(radius=self.radius)
        bend.connect(port='o1', destination=mzi.ports['o2'])
        self.component.add_port('o1', port=mzi.ports['o1'])
        self.component.add_port('o2', port=bend.ports['o2'])

        return self.component

if __name__ == '__main__':
    # Instantiate the class with a specific radius
    mzi_bend = MZIWithBend(radius=50)
    
    # Create the component
    c = mzi_bend.create_component()
    
    # Add fiber array for routing
    c = gf.routing.add_fiber_array(c)
    
    # Export the design to a GDS file
    c.write_gds("input_oop.gds")
    
    # Display the component in a viewer
    c.show()
    
    # Plot the component layout
    c.plot()
```

```python
import gdsfactory as gf

# Function to create a new MZI component and connect it to a bend
def create_mzi_with_bend(radius: float) -> gf.Component:
    """
    Creates a new component consisting of an MZI with a circular bend.
    
    Args:
        radius: Radius of the circular bend.
    
    Returns:
        A gdsfactory Component with an MZI and a bend.
    """
    c = gf.Component()
    
    # Create Mach-Zehnder Interferometer (MZI)
    mzi = c << gf.components.mzi(delta_length=10)
    
    # Create a new bend component
    bend = gf.components.bend_circular(radius=radius)
    bend_ref = c << bend
    
    # Connect the bend to the MZI
    bend_ref.connect(port='o1', destination=mzi.ports['o2'])
    
    # Add ports to the component
    c.add_port('o1', port=mzi.ports['o1'])
    c.add_port('o2', port=bend_ref.ports['o2'])
    
    return c

def main():
    # Create the component using the corrected function
    c = create_mzi_with_bend(radius=50)
    
    # Add fiber array for routing
    c = gf.routing.add_fiber_array(c)
    
    # Export the design to a GDS file
    c.write_gds("input_lambda_corrected.gds")
    
    # Display the component in a viewer
    c.show()
    
    # Plot the component layout
    c.plot()

if __name__ == '__main__':
    main()
```

```python
import gdsfactory as gf
import matplotlib.pyplot as plt

# Define the custom straight component
def straight(length=10, width=1, layer=(1, 0)):
    c = gf.Component()
    # Add a polygon representing the waveguide
    c.add_polygon([(0, 0), (length, 0), (length, width), (0, width)], layer=layer)
    # Add ports for connection
    c.add_port(
        name="o1", center=[0, width / 2], width=width, orientation=180, layer=layer
    )
    c.add_port(
        name="o2", center=[length, width / 2], width=width, orientation=0, layer=layer
    )
    return c

# Create a component to use the custom straight components
c = gf.Component()

# Instantiate and position the waveguides
wg1 = c << straight(length=6, width=2.5, layer=(1, 0))
wg2 = c << straight(length=6, width=2.5, layer=(2, 0))
wg3 = c << straight(length=15, width=2.5, layer=(3, 0))

# Move and rotate waveguides for better layout
wg2.dmove((0, 10))
wg2.drotate(10)

wg3.dmove((0, 20))
wg3.drotate(15)

# Plot the layout of the waveguides
c.plot()

# Define other components
gc = gf.components.grating_coupler_elliptical(
    polarization='te', taper_length=16.6, taper_angle=40, wavelength=1.554, 
    fiber_angle=15, grating_line_width=0.343, neff=2.638, nclad=1.443, 
    n_periods=30, big_last_tooth=False, layer_slab='SLAB150', slab_xmin=-1, 
    slab_offset=2, spiked=True, cross_section='strip'
)

gc1 = gf.components.grating_coupler_elliptical(
    polarization='te', taper_length=16.6, taper_angle=40, wavelength=1.554, 
    fiber_angle=15, grating_line_width=0.343, neff=2.638, nclad=1.443, 
    n_periods=30, big_last_tooth=False, layer_slab='SLAB150', slab_xmin=-1, 
    slab_offset=2, spiked=True, cross_section='strip'
)

bend180 = gf.components.bezier(
    control_points=((5, 0), (5, 0), (5, 0), (7, 0)), npoints=10, 
    with_manhattan_facing_angles=True, cross_section='strip', 
    allow_min_radius_violation=False
)

mzi = gf.components.coupler_ring(
    gap=0.2, radius=5, length_x=0.5, bend='bend_euler', 
    straight='straight', cross_section='strip', length_extension=0
)

wg = gf.components.crossing_from_taper()

# Define the symbol to component mapping
symbol_to_component = {
    "A": (bend180, 'o1', 'o2'),
    "B": (bend180, 'o2', 'o1'),
    "C": (mzi, 'o1', 'o2'),
    "D": (wg, 'o1', 'o2'),
    "F": (gc, 'o1', 'o2'),
    "G": (gc1, 'o1', 'o2')
}

# Create the component sequence
s = "CDC"
c_seq = gf.components.component_sequence(sequence=s, symbol_to_component=symbol_to_component)

# Plot the sequence
c_seq.plot()

# Let's keep wg1 in place on the bottom, and connect the other straights to it.
# To do that, on wg2 we'll grab the "o1" port and connect it to the "o2" on wg1:
wg2.connect("o1", wg1.ports["o2"], allow_layer_mismatch=True)

# Next, on wg3 let's grab the "o1" port and connect it to the "o2" on wg2:
wg3.connect("o1", wg2.ports["o2"], allow_layer_mismatch=True)

c.plot()
```

```python
import gdsfactory as gf
from gdsfactory.labels.add_label_yaml import add_label_json

# Function to create a spiral component with a given length
def create_spiral_with_length(length):
    c = gf.components.spiral(length=length)
    c.info["measurement"] = "optical_loopback2"
    c = add_label_json(c)
    return c

# List of lengths to sweep over
lengths = [10, 50, 100, 200]

# Create a list of spiral components with different lengths
sweep = [create_spiral_with_length(length) for length in lengths]

# Pack the components together
m = gf.pack(sweep)

# Plot and show each packed component
for c in m:
    c.plot()
    c.show()
```

```python
import gdsfactory as gf
import gdsfactory.schematic as gt
splitter = gf.components.mmi1x2()
n = 5
dx = 300
dy = 300
s = gt.Schematic()

for col in range(n):
    rows = 2**col
    for row in range(rows):
        s.add_instance(f"s{col}{row}", gt.Instance(component=splitter))
        s.add_placement(
            f"s{col}{row}", gt.Placement(x=col * dx, y=(row - rows / 2) * dy)
        )
        if col < n - 1:
            s.add_net(gt.Net(p1=f"s{col}{row},o2", p2=f"s{col+1}{2*row},o1"))
            s.add_net(gt.Net(p1=f"s{col}{row},o3", p2=f"s{col+1}{2*row+1},o1"))


g = s.plot_netlist()
```

```python
n = 2**3
splitter = gf.components.splitter_tree(noutputs=n, spacing=(50, 50))
dbr_array = gf.components.array(
    component=gf.c.dbr, rows=n, columns=1, spacing=(0, 3), centered=True
)
s = gt.Schematic()
s.add_instance("s", gt.Instance(component=splitter))
s.add_instance("dbr", gt.Instance(component=dbr_array))
s.add_placement("s", gt.Placement(x=10, y=10))
s.add_placement("dbr", gt.Placement(x=300, y=0))

for i in range(n):
    s.add_net(
        gt.Net(
            p1=f"s,o2_2_{i+1}",
            p2=f"dbr,o1_{i+1}_1",
            name="splitter_to_dbr",
            settings=dict(radius=5, sort_ports=True),
        )
    )

g = s.plot_netlist()
```
